## Aula 5 - Machine Learning


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

pd.set_option("display.precision", 2)
pd.set_option('display.float_format', lambda x: '%.2f' % x)
sns.set()

dados_vendas_censo = pd.read_csv("dados_vendas_censo.csv")
dados_vendas_censo = dados_vendas_censo[['Rua', 'Bairro', 'Cidade', 'Metragem', 'Quartos', 'Banheiros', 'Vagas',
       'Valor_anuncio', 'Valor_mm', 'Valor_m2', 'latitude', 'longitude', 'Situacao_setor', 'V001',
       'V002', 'V003', 'V004', 'V005', 'V006', 'V007', 'V008', 'V009', 'V010',
       'V011', 'V012']].dropna()
dados_vendas_censo
```

```python
plt.figure(figsize = (10, 8))
ax = sns.boxplot(data = dados_vendas_censo, y = "Valor_anuncio")
plt.show()
```

```python
dados_vendas_censo.query("Valor_anuncio > 30000000 | Valor_anuncio < 100000") # dá para excluir esses valores
```

```python
dados_vendas_censo.drop(dados_vendas_censo.query("Valor_anuncio > 30000000 | Valor_anuncio < 100000").index, inplace = True) # excluindo os valores extremos
plt.figure(figsize=(10, 8))
ax = sns.boxplot(data = dados_vendas_censo, y="Valor_anuncio")
plt.show()
```

```python
plt.figure(figsize = (10, 10))
sns.scatterplot(data = dados_vendas_censo, x = "Valor_anuncio", y = "Valor_m2") # o metro quadrado do imóvel tem relação com o valor de anúncio
```

```python
dados_vendas_censo.corr() # analisando a correlação
```

```python
plt.figure(figsize=(23, 10))
#https://www.tylervigen.com/spurious-correlations
mask = np.triu(np.ones_like(dados_vendas_censo.corr(), dtype=bool))
heatmap = sns.heatmap(dados_vendas_censo.corr(), mask=mask, vmin=-1, vmax=1, annot=True, cmap='BrBG')
heatmap.set_title('Correlação', fontdict={'fontsize':20}, pad=25);
```

```python
from sklearn.linear_model import LinearRegression # vamos trabalhar com regressão
from sklearn.model_selection import train_test_split

X = dados_vendas_censo[["Metragem"]] # variável x
Y = dados_vendas_censo["Valor_anuncio"] # variável y

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 54) # separando em valores de treino e de teste
lr = LinearRegression() 
```

```python
X_train # treino em x
```

```python
Y_train # treino em y
```

```python
lr.fit(X_train, Y_train) # treinando o modelo
```

```python
Y_predict_test = lr.predict(X_test) # testando o modelo
```

```python
Y_test
```

```python
Y_predict_train = lr.predict(X_train)
```

```python
Y_train # teste em y
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

mean_absolute_error(Y_test, Y_predict_test) # quanto maior o número, pior é o resultado
```

```python
mean_absolute_error(Y_train, Y_predict_train) # quanto maior o número, pior é o resultado
```

```python
X = dados_vendas_censo[['Metragem', 'Quartos', 'Banheiros', 'Vagas','V001','V007', 'V009']] # adicionando as variáveis

Y = dados_vendas_censo["Valor_anuncio"]

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.25, random_state = 54)
lr.fit(X_train, Y_train)
Y_predict_test = lr.predict(X_test)
Y_predict_train = lr.predict(X_train)
mae_test = mean_absolute_error(Y_test, Y_predict_test)
r2_test = r2_score(Y_test, Y_predict_test)
mae_train = mean_absolute_error(Y_train, Y_predict_train)
r2_train = r2_score(Y_train, Y_predict_train)
```
